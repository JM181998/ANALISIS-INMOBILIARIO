```python
#Normalizamos los nombres de las columnas
def normalizar_nombres(nombres):
    nombres = nombres.lower().replace(" ", "_") 
    nombres = nombres.replace('á', 'a').replace('é', 'e').replace('í', 'i').replace('ó', 'o').replace('ü', 'u').replace('ú', 'u')
    nombres = nombres.replace('Á', 'a').replace('É', 'e').replace('Í', 'i').replace('Ó', 'o').replace('Ú', 'u')
    return nombres
df.columns = [normalizar_nombres(col) for col in df.columns]
```

```python
# Limpieza de la columna precio
def limpiar_precio(x):
    try:
        if 'A consultar' in x or 'Error al procesar el precio' in x:
            return None
        if '€/día' in x:
            return float(x.replace(' €/día', '').replace('€', '').replace(' ', '').replace('.', '').replace(',', '.')) * 30
        if '€/sem' in x:
            return float(x.replace(' €/sem', '').replace('€', '').replace(' ', '').replace('.', '').replace(',', '.')) * 4
        return float(x.replace(' €/mes', '').replace('€', '').replace(' ', '').replace('.', '').replace(',', '.'))
    except ValueError:
        return None

df['precio'] = df['precio'].apply(limpiar_precio)
df['precio'] = df['precio'].astype(float)
```

```python
# Aplicamos limpieza a habitaciones según lo encontrado
def limpiar_habitaciones(x):
    try:
        x = float(x)
        if x == 90:
            return 3
        elif x == 80:
            return 1
        elif x == 22:
            return 2
        elif x == 20:
            return 20
        elif x == 21:
            return None
        elif x > 20 or not x.is_integer():
            return None
        return int(x)
    except ValueError:
        return None

df['habitaciones'] = df['habitaciones'].apply(limpiar_habitaciones).astype('Int64')
```

```python
# Vamos a aplicar una logica común ya que este dato en la página da tres posibilidades:
# Escribir la caracteristica concreta o especificar que si, lo hemos reducido a "Si" por simplificar
# Escribir que no, implica confirmación de que no se da esa caracteristica en dicho inmueble por lo que la tratamos como "No"
# La tercera opción es que devolviera la clave valor vacía, lo que significa que no se especifica, de esta forma no nos inventamos información
def limpiar_columnas_si_no(valor, valores_unicos):
    if pd.isna(valor) or valor.strip() == "":
        return "Sin especificar"
    if valor == "Sin especificar":
        return valor
    elif "no" in valor.lower().split():
        return "No"
    else:
        return "Si"

# Iterar sobre cada columna conflictiva y limpiar los valores
for col in columnas_si_no:
    valores_unicos = df[col].dropna().unique()
    df[col] = df[col].apply(lambda x: limpiar_columnas_si_no(x, valores_unicos))
```

```python
def clean_gastos_de_comunidad(value):
    if pd.isna(value):
        return "Sin especificar"
    value = value.lower().strip()
    if any(x in value for x in ['incluido', 'incluida', 'incluídos', 'incluids', 'inlcuidos', 'incuídos', 'no tiene', "sin comunidad", "no aplicable", "no hay",
                                'a cargo de la propiedad', 'se repercute al inquilino optante']):
        return "No"
    if any(x in value for x in ['mas de 100', 'más de 100', 'mas 150']):
        return ">100"
    if any(x in value for x in ['entre 10 y 20']):
        return "10-20"
    if any(x in value for x in ['entre 20 y 40']):
        return "20-40"
    if any(x in value for x in ['entre 40 y 60']):
        return "40-60"
    if any(x in value for x in ['entre 60 y 80']):
        return "60-80"
    if any(x in value for x in ['entre 80 y 100']):
        return "80-100"
    if any(x in value for x in ['al mes.', 'a consultar', '30 euros de comunidad', 'alquiler de septiembre a junio', '-', 'si']):
        return "Sin especificar"
    try:
        # Try to convert to float and categorize
        num_value = float(value.replace('€','').replace(',','.').strip())
        if num_value <= 20:
            return "<20"
        elif num_value <= 40:
            return "20-40"
        elif num_value <= 60:
            return "40-60"
        elif num_value <= 80:
            return "60-80"
        elif num_value <= 100:
            return "80-100"
        else:
            return ">100"
    except ValueError:
        return "Sin especificar"

# Apply the cleaning function to the column
df['gastos_de_comunidad'] = df['gastos_de_comunidad'].apply(clean_gastos_de_comunidad)
```

```python
# Creo que pisciina puede ser una variable extra a tener en cuenta, asi que la limpio con mas detalle
def clean_piscina(value):
    value = value.lower()
    if any(x in value for x in ['si', 'yes']):
        return "Si"
    if any(x in value for x in ['no']):
        return "No"
    if any(x in value for x in ['propia', 'private','jacuzzi']):
        return "Propia"
    if any(x in value for x in ['comunitaria', 'community','comunitarias','climatizada','infantil','interior']):
        return "Comunitaria"
    else:
        return "Sin especificar"

df['piscina'] = df['piscina'].apply(clean_piscina)
```
